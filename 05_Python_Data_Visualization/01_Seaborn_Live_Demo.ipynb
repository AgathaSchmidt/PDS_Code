```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
diamonds["cut"].nunique()
```

```python
diamonds["cut"].unique()
```

```python
diamonds.describe()
```

```python
sns.distplot(diamonds["price"], bins=10, kde=False)
plt.show()
```

```python
sns.jointplot(x="carat", y="price", data=diamonds, kind="reg")
plt.show()
```

```python
sns.pairplot(diamonds)
```

```python
sns.barplot(x="cut", y="price", data=diamonds, estimator=np.mean)
```

```python
sns.countplot(x="cut", data=diamonds)
```

```python
sns.boxplot(x="cut", y="price", data=diamonds)
```

```python
sns.violinplot(x="cut", y="price", data=diamonds)
```

```python
sns.violinplot(x="cut", y="price", hue="color", 
               data=diamonds[(diamonds["color"]=="G") | (diamonds["color"] == "E")], split=True)
```

```python
sns.heatmap(diamonds.corr())
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="carat", y="price", hue="color", 
           data=diamonds[(diamonds["color"]=="G") | (diamonds["color"] == "E")],
           palette="coolwarm")
```
