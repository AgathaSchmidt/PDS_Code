```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r')
plt.show()
```

```python
plt.plot(x, y, 'r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6), dpi=200)

ax.plot(x, y, 'r', label="Sine")

for k in range(1,10):
    if k != 5:
        ax.plot(x, (k/5)*y, 'r--', linewidth=0.3*(1-((k/5)-1)**2))
        
y2 = np.cos(x)
ax.plot(x, y2, 'b', label="Cosine")

for k in range(1,10):
    if k != 5:
        ax.plot(x, (k/5)*y2, 'b--', linewidth=0.3*(1-((k/5)-1)**2))

ax.set_xlim([-2*np.pi, 2*np.pi])

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Sine")

ax.legend()

plt.show()
```

```python
x = np.random.randn(100)
y = 2*x+np.random.randn(100)
z = np.abs(x**2+2*y**3+np.random.randn(100))
c = np.abs(z*y+np.random.randn(100))
```

```python
data = np.array([x,y,z,c]).transpose()
```

```python
df = pd.DataFrame(data=data, columns=["x", "y", "z", "c"])
df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6), dpi=200)

scatter = ax.scatter(x=df["x"], y=df["y"], 
                         s=df["z"], c=df["c"])

ax.set_xlabel("X")
ax.set_ylabel("Y")

cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label("C")

size_handles, size_labels = scatter.legend_elements(prop="sizes", alpha=0.5)
ax.legend(size_handles, size_labels, loc="lower right", title="Z")

plt.show()
```

```python
df = pd.DataFrame(data=data, columns=["x", "y", "z", "c"])
df_list = []

df_list.append(df[df["z"]>=50])
df_list.append(df[df["z"]<50])

markers = ['o', '*']

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6), dpi=200)

for i, df in enumerate(df_list):
    scatter = ax[i].scatter(x=df["x"], y=df["y"], s=df["z"], c=df["c"], marker=markers[i])
    
    ax[i].set_xlabel("X")
    ax[i].set_ylabel("Y")

    scatter.set_clim([0,4000])
    cbar = fig.colorbar(scatter, ax=ax[i])
    cbar.set_label("C")

    size_handles, size_labels = scatter.legend_elements(prop="sizes", alpha=0.5)
    ax[i].legend(size_handles, size_labels, loc="lower right", title="Z")
```
